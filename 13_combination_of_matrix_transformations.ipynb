```python
import numpy as np
import matplotlib.pyplot as plt

# Define basis vectors
e1 = np.array([1, 0])
e2 = np.array([0, 1])

# Define transformation matrices
A1 = np.array([[0, -1], [1, 0]])  # 90 degree rotation matrix
A2 = np.array([[-1, 0], [0, 1]])  # Vertical reflection matrix

# Apply transformations
e1_prime = A1 @ e1  # Apply A1 to e1
e2_prime = A1 @ e2  # Apply A1 to e2

# Now apply A2 to the transformed vectors
e1_double_prime = A2 @ e1_prime
e2_double_prime = A2 @ e2_prime

# Plotting the results
plt.figure(figsize=(6,6))

# Original basis vectors (e1 and e2)
plt.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color='black', label='e1 (Original)')
plt.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='black', label='e2 (Original)')

# Transformed vectors (after A1 and A2)
plt.quiver(0, 0, e1_prime[0], e1_prime[1], angles='xy', scale_units='xy', scale=1, color='blue', label='e1\' (After A1)')
plt.quiver(0, 0, e2_prime[0], e2_prime[1], angles='xy', scale_units='xy', scale=1, color='blue', label='e2\' (After A1)')

# Transformed vectors after A2
plt.quiver(0, 0, e1_double_prime[0], e1_double_prime[1], angles='xy', scale_units='xy', scale=1, color='red', label='e1\'\' (After A2)')
plt.quiver(0, 0, e2_double_prime[0], e2_double_prime[1], angles='xy', scale_units='xy', scale=1, color='red', label='e2\'\' (After A2)')

# Plot the reference square (original)
plt.plot([0, e1[0]], [0, e1[1]], color='black')  # e1
plt.plot([0, e2[0]], [0, e2[1]], color='black')  # e2
plt.plot([e1[0], e1[0] + e2[0]], [e1[1], e1[1] + e2[1]], color='black')  # e1 + e2
plt.plot([e2[0], e2[0] + e1[0]], [e2[1], e2[1] + e1[1]], color='black')  # e2 + e1

# Plot the transformed parallelogram (after reflection and rotation)
plt.plot([0, e1_prime[0]], [0, e1_prime[1]], color='blue', linestyle='--')  # e1' line
plt.plot([0, e2_prime[0]], [0, e2_prime[1]], color='blue', linestyle='--')  # e2' line
plt.plot([e1_prime[0], e1_prime[0] + e2_prime[0]], [e1_prime[1], e1_prime[1] + e2_prime[1]], color='blue', linestyle='--')  # e1' + e2'
plt.plot([e2_prime[0], e2_prime[0] + e1_prime[0]], [e2_prime[1], e2_prime[1] + e1_prime[1]], color='blue', linestyle='--')  # e2' + e1'

plt.plot([0, e1_double_prime[0]], [0, e1_double_prime[1]], color='red', linestyle='--')  # e1'' line
plt.plot([0, e2_double_prime[0]], [0, e2_double_prime[1]], color='red', linestyle='--')  # e2'' line
plt.plot([e1_double_prime[0], e1_double_prime[0] + e2_double_prime[0]], [e1_double_prime[1], e1_double_prime[1] + e2_double_prime[1]], color='red', linestyle='--')  # e1'' + e2''
plt.plot([e2_double_prime[0], e2_double_prime[0] + e1_double_prime[0]], [e2_double_prime[1], e2_double_prime[1] + e1_double_prime[1]], color='red', linestyle='--')  # e2'' + e1''

# Formatting the plot
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Composition of Rotation and Reflection')
plt.show()
```
