```python
import numpy as np
import matplotlib.pyplot as plt

# Define the original unit vectors (basis vectors)
e1 = np.array([1, 0])  # Unit vector along the x-axis
e2 = np.array([0, 1])  # Unit vector along the y-axis

# Define the transformation matrix
M = np.array([[2, 3], [10, 1]])

# Apply the transformation to the unit vectors
e1_transformed = M @ e1
e2_transformed = M @ e2

# Plotting
plt.figure(figsize=(8, 6))

# Plot the original unit vectors (blue)
plt.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color='blue', label=r'Original $e_1 = [1, 0]^T$')
plt.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='blue', label=r'Original $e_2 = [0, 1]^T$')

# Plot the transformed vectors (red)
plt.quiver(0, 0, e1_transformed[0], e1_transformed[1], angles='xy', scale_units='xy', scale=1, color='red', label=r'Transformed $M \cdot e_1 = [2, 10]^T$')
plt.quiver(0, 0, e2_transformed[0], e2_transformed[1], angles='xy', scale_units='xy', scale=1, color='red', label=r'Transformed $M \cdot e_2 = [3, 1]^T$')

# Set plot labels and grid
plt.xlim(-1, 4)
plt.ylim(-1, 12)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Transformation of Unit Vectors by Matrix M')

# Show legend
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for a (price of apples)
a_values = np.linspace(0, 5, 100)

# Define the corresponding b values from both equations
b_values_1 = (8 - 2 * a_values) / 3  # From 2a + 3b = 8
b_values_2 = 13 - 10 * a_values  # From 10a + b = 13

# Create the plot
plt.figure(figsize=(8, 6))

# Plot both lines
plt.plot(a_values, b_values_1, label=r'$2a + 3b = 8$', color='blue')
plt.plot(a_values, b_values_2, label=r'$10a + b = 13$', color='red')

# Set plot labels
plt.title('Graph of the System of Equations')
plt.xlabel('Price of Apples (a)')
plt.ylabel('Price of Bananas (b)')

# Set limits for better viewing
plt.xlim(0, 5)
plt.ylim(0, 5)

# Show the intersection point (solution to the system)
plt.plot(3, 1, 'go')  # This is the intersection point (a=3, b=1)
plt.text(3, 1, '  (3, 1)', fontsize=12)

# Display the legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
