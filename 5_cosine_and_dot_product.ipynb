```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image_path = './assets/images/cosine_rule_vector.png'

show_image(image_path, 200)
```

```python
# Given sides and angle (in degrees)
a = 5
b = 7
theta_deg = 60  # Angle between sides a and b
theta_rad = np.radians(theta_deg)  # Convert to radians

# Compute the third side using the Cosine Rule
c = np.sqrt(a**2 + b**2 - 2 * a * b * np.cos(theta_rad))

# Triangle coordinates
A = (0, 0)  # First point
B = (b, 0)  # Second point along x-axis
C = (a * np.cos(theta_rad), a * np.sin(theta_rad))  # Third point using polar coordinates

# Plot the triangle
plt.figure(figsize=(6,6))
plt.plot([A[0], B[0]], [A[1], B[1]], 'bo-')  # Line AB
plt.plot([B[0], C[0]], [B[1], C[1]], 'ro-')  # Line BC
plt.plot([C[0], A[0]], [C[1], A[1]], 'go-')  # Line CA

# Annotate points
plt.text(A[0], A[1], ' A', fontsize=12, verticalalignment='bottom', horizontalalignment='right')
plt.text(B[0], B[1], ' B', fontsize=12, verticalalignment='bottom', horizontalalignment='left')
plt.text(C[0], C[1], ' C', fontsize=12, verticalalignment='bottom', horizontalalignment='left')

# Annotate side lengths
plt.text((A[0]+B[0])/2, (A[1]+B[1])/2, f'{b}', fontsize=12, color='blue')
plt.text((B[0]+C[0])/2, (B[1]+C[1])/2, f'{c:.2f}', fontsize=12, color='red')
plt.text((C[0]+A[0])/2, (C[1]+A[1])/2, f'{a}', fontsize=12, color='green')

# Title and grid
plt.title("Triangle Visualization Using Cosine Rule")
plt.grid()
plt.xlim(-1, b + 2)
plt.ylim(-1, a + 2)
plt.gca().set_aspect('equal')  # Keep aspect ratio square


plt.show()
```
