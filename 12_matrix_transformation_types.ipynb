```python
img_path = "./assets/images/12_4.png"
show_image(img_path, 200)

# Identity matrix
I = np.array([[1, 0], [0, 1]])

# Define vectors
v = np.array([3, 2])
e1 = np.array([1, 0])
e2 = np.array([0, 1])

# Apply the identity matrix (no change)
v_transformed = I @ v

# Plot the original and transformed vectors along with reference basis vectors
plt.figure(figsize=(6,6))
# Original vectors
plt.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color='black', label='e1 (Original)')
plt.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='black', label='e2 (Original)')
# Transformed vectors
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original Vector')
plt.quiver(0, 0, v_transformed[0], v_transformed[1], angles='xy', scale_units='xy', scale=1, color='red', label='Transformed Vector')

# Closing the square (original vectors form a unit square)
plt.plot([0, e1[0]], [0, e1[1]], color='black')  # e1
plt.plot([0, e2[0]], [0, e2[1]], color='black')  # e2
plt.plot([e1[0], e1[0] + e2[0]], [e1[1], e1[1] + e2[1]], color='black')  # e1 + e2
plt.plot([e2[0], e2[0] + e1[0]], [e2[1], e2[1] + e1[1]], color='black')  # e2 + e1

# Formatting the plot
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Identity Transformation')
plt.show()
```

```python
img_path = "./assets/images/12_9.png"
show_image(img_path, 200)

# Shear matrix
Shear = np.array([[1, 1], [0, 1]])

# Apply the shear matrix to the vector
v_transformed = Shear @ v

# Create the transformed reference basis vectors
e1_transformed = Shear @ e1
e2_transformed = Shear @ e2

# Plot the original and transformed vectors along with the reference basis vectors
plt.figure(figsize=(6,6))

# Original vectors (Basis vectors)
plt.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color='black', label='e1 (Original)')
plt.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='black', label='e2 (Original)')

# Transformed vectors (Sheared Vector)
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original Vector')
plt.quiver(0, 0, v_transformed[0], v_transformed[1], angles='xy', scale_units='xy', scale=1, color='red', label='Transformed Vector')

# Closing the original square (unit square formed by basis vectors)
plt.plot([0, e1[0]], [0, e1[1]], color='black')  # e1
plt.plot([0, e2[0]], [0, e2[1]], color='black')  # e2
plt.plot([e1[0], e1[0] + e2[0]], [e1[1], e1[1] + e2[1]], color='black')  # e1 + e2
plt.plot([e2[0], e2[0] + e1[0]], [e2[1], e2[1] + e1[1]], color='black')  # e2 + e1

# Closing the transformed parallelogram (sheared unit square)
plt.plot([0, e1_transformed[0]], [0, e1_transformed[1]], color='red', linestyle='--')  # e1 transformed
plt.plot([0, e2_transformed[0]], [0, e2_transformed[1]], color='red', linestyle='--')  # e2 transformed
plt.plot([e1_transformed[0], e1_transformed[0] + e2_transformed[0]], [e1_transformed[1], e1_transformed[1] + e2_transformed[1]], color='red', linestyle='--')  # e1_transformed + e2_transformed
plt.plot([e2_transformed[0], e2_transformed[0] + e1_transformed[0]], [e2_transformed[1], e2_transformed[1] + e1_transformed[1]], color='red', linestyle='--')  # e2_transformed + e1_transformed

# Formatting the plot
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Shear Transformation with Parallelogram Shift')
plt.show()


```

```python
img_path = "./assets/images/12_10.png"
show_image(img_path, 200)

# Rotation matrix for 45 degrees counterclockwise
theta = np.radians(45)  # Convert to radians
R_theta = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])

# Apply the rotation matrix
v_transformed = R_theta @ v

# Plot the original and transformed vectors
plt.figure(figsize=(5,5))
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original Vector')
plt.quiver(0, 0, v_transformed[0], v_transformed[1], angles='xy', scale_units='xy', scale=1, color='red', label='Transformed Vector')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
