```python
import numpy as np
from fractions import Fraction

# Define matrix A with fractions
A = np.array([
    [Fraction(1, 1), Fraction(3, 2), Fraction(1, 2)],
    [Fraction(0, 1), Fraction(1, 1), Fraction(1, 1)],
    [Fraction(2, 1), Fraction(8, 1), Fraction(13, 1)]
])

# Calculate the inverse using np.linalg.inv
A_inv = np.linalg.inv(A.astype(float))

# Print A_inv using fractions
def display_fractions(matrix):
    rows, cols = matrix.shape
    for i in range(rows):
        row = matrix[i]
        print("[" + ", ".join([f"{Fraction(x).limit_denominator()}" for x in row]) + "]")

# Display A_inv as fractions
display_fractions(A_inv)
```
