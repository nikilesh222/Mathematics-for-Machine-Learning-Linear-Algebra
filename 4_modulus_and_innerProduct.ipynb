```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Markdown
import matplotlib.image as mpimg
```

```python
# Section 1: Introduction
display(Markdown("""

### Understanding Vectors, Length, and the Dot Product

This notebook explores fundamental vector operations: vector addition, scaling, length calculation, and the dot product.

"""))
```

```python
# Section 2: Visualizing Vectors in 2D
vector_r = np.array([3, 2])  # Vector r
vector_s = np.array([-1, 2])  # Vector s

fig, ax = plt.subplots()
ax.quiver(0, 0, vector_r[0], vector_r[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector r')
ax.quiver(0, 0, vector_s[0], vector_s[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector s')
ax.set_xlim(-2, 5)
ax.set_ylim(-1, 4)
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)
ax.grid()
ax.legend()
plt.title("Vector Visualization")
plt.show()

display(Markdown("""
### Visualizing Vectors in Space

The vectors **r** and **s** are shown in 2D space. Vector **r** is 3 units along the x-axis and 2 units along the y-axis. Vector **s** is moving back 1 unit along the x-axis and up 2 units along the y-axis.

Here, you can see how vector addition works geometrically: placing vector **s** at the tip of **r** gives you the vector **r + s**.
"""))
```

```python
# Section 3: Computing the Length of a Vector
def vector_length(v):
    return np.sqrt(np.sum(v**2))

length_r = vector_length(vector_r)
display(Markdown(f"""
### Computing the Length of a Vector

The length (magnitude) of vector **r** is computed using the formula:
\[ |r| = \sqrt{{a^2 + b^2}} \]

For **r = (3,2)**:
\[ |r| = \sqrt{{3^2 + 2^2}} = {length_r:.2f} \]
"""))

image_path = 'assets/images/vector_length_or_magnitude.png'

show_image(image_path, 150)
```

```python
# Section 4: Dot Product of Two Vectors
def dot_product(v1, v2):
    return np.dot(v1, v2)

dot_rs = dot_product(vector_r, vector_s)
display(Markdown(f"""
### Dot Product of Two Vectors

The dot product formula is:
\[ r \cdot s = r_x s_x + r_y s_y \]

For **r = (3,2)** and **s = (-1,2)**:
\[ 3 \times (-1) + 2 \times 2 = {dot_rs} \]

The result of the dot product is **{dot_rs}**, which is a scalar value.
"""))

image_path = './assets/images/vector_dot_product.png'

show_image(image_path, 150)
```

```python
# Section 5: Properties of the Dot Product
def check_commutativity(v1, v2):
    return np.dot(v1, v2) == np.dot(v2, v1)

def check_distributivity(v1, v2, v3):
    return np.allclose(np.dot(v1, v2 + v3), np.dot(v1, v2) + np.dot(v1, v3))

def check_associativity(v1, v2, scalar):
    return np.allclose(np.dot(v1, scalar * v2), scalar * np.dot(v1, v2))

vector_t = np.array([1, 3])  # A third vector
a_scalar = 2  # A scalar value

commutative = check_commutativity(vector_r, vector_s)
distributive = check_distributivity(vector_r, vector_s, vector_t)
associative = check_associativity(vector_r, vector_s, a_scalar)

display(Markdown(f"""
### Properties of the Dot Product

1. **Commutativity:** Does **r  s** equal **s  r**?  {commutative}
2. **Distributivity over addition:** Does **r  (s + t) = r  s + r  t**?  {distributive}
3. **Associativity over scalar multiplication:** Does **r  (a * s) = a * (r  s)**?  {associative}
"""))
```

```python
# Section 6: Relationship Between Dot Product and Magnitude
display(Markdown("""
### Relationship Between the Dot Product and Vector Length

If we take the dot product of a vector with itself, we get its squared magnitude:
\[ r \cdot r = |r|^2 \]

So, taking the square root of the dot product of a vector with itself gives its length.
"""))

self_dot_product = dot_product(vector_r, vector_r)
computed_length = np.sqrt(self_dot_product)

display(Markdown(f"""
For **r = (3,2)**:
\[ r \cdot r = 3^2 + 2^2 = {self_dot_product} \]

Thus, the magnitude \(|r| = \sqrt{{{self_dot_product}}} = {computed_length:.2f}\), which matches our earlier calculation.
"""))
```
