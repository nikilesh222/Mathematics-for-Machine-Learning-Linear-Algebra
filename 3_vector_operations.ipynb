```python
# In[1]: Import Necessary Libraries
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
# In[2]: Section 1 - Introduction to Vectors
# Display a brief explanation of vectors in the context of physics and data science.

from IPython.display import display, Markdown

display(Markdown("""
# Understanding Vectors

A vector can be thought of as an object that moves us around space. This could be a physical space in physics or even a space of data in data science.

In data science, a vector represents attributes of an object, for example:

- A house can be represented by a vector: [120 m² (floor area), 2 bedrooms, 1 bathroom, 150,000 Euros (price)].

### Key Vector Operations:
1. **Addition**: Adding two vectors together.
2. **Scalar multiplication**: Multiplying a vector by a scalar.
"""))
```

```python
# In[3]: Section 2 - Geometric Representation of Vectors
# Visualizing vectors in 2D space with matplotlib.

plt.figure(figsize=(6, 6))
plt.quiver(0, 0, 3, 2, angles='xy', scale_units='xy', scale=1, color="blue", label="Vector r")
plt.quiver(3, 2, -1, 2, angles='xy', scale_units='xy', scale=1, color="red", label="Vector s")
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title("Geometric Representation of Vectors r and s")

plt.show()

display(Markdown("""
### Visualizing Vectors in Space
The vectors r and s are shown in 2D space. Vector **r** is 3 units along the x-axis and 2 units along the y-axis. Vector **s** is moving back 1 unit along the x-axis and up 2 units along the y-axis.

Here, you can see how vector addition works geometrically: placing vector **s** at the tip of **r** gives you the vector **r + s**.
"""))
```

```python
# In[4]: Section 3 - Vector Addition and Commutativity
# Let's calculate the sum of two vectors using component-wise addition.

# Define the components of the vectors
r = np.array([3, 2])  # Vector r = 3i + 2j
s = np.array([-1, 2])  # Vector s = -1i + 2j

# Add the vectors
r_plus_s = r + s

# Show the result
display(Markdown(f"""
### Vector Addition: r + s
Given vectors:
- r = [3, 2]
- s = [-1, 2]

The sum of vectors **r** and **s** is: **r + s = {r_plus_s}**
"""))

plt.figure(figsize=(6, 6))
plt.quiver(0, 0, 3, 2, angles='xy', scale_units='xy', scale=1, color="blue", label="Vector r")
plt.quiver(3, 2, -1, 2, angles='xy', scale_units='xy', scale=1, color="red", label="Vector s")
plt.quiver(0, 0, 2, 4, angles='xy', scale_units='xy', scale=1, color="green", label="r + s (Result)")

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title("Vector Addition (r + s)")

plt.show()

display(Markdown("""
### Commutativity of Vector Addition
Notice that it doesn't matter if you add **r + s** or **s + r**. The result will always be the same. This is known as **commutative** property of vector addition.
"""))
```

```python
# In[5]: Section 4 - Scalar Multiplication
# Visualize how scalar multiplication scales the vector.

# Scalar multiplication of r by 2
scaled_r = 2 * r

# Scalar multiplication of r by -1
neg_r = -1 * r

display(Markdown("""
### Scalar Multiplication
We can scale a vector by multiplying it by a scalar (a number). For example:

- **2r** means we multiply each component of **r** by 2, resulting in a scaled vector that is twice as long.
- **-r** means we scale **r** by -1, which points in the opposite direction.

Let's visualize these operations.
"""))

# Plot the vectors
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, 3, 2, angles='xy', scale_units='xy', scale=1, color="blue", label="Vector r")
plt.quiver(0, 0, scaled_r[0], scaled_r[1], angles='xy', scale_units='xy', scale=1, color="purple", label="2r")
plt.quiver(0, 0, neg_r[0], neg_r[1], angles='xy', scale_units='xy', scale=1, color="orange", label="-r")
plt.xlim(-1, 7)
plt.ylim(-3, 5)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title("Scalar Multiplication (2r and -r)")

plt.show()
```

```python
# In[6]: Section 5 - Using a Coordinate System
# Let's define a simple coordinate system with i and j as basis vectors.

display(Markdown("""
### Coordinate System and Basis Vectors

We define a coordinate system with two basis vectors **i** (unit vector along the x-axis) and **j** (unit vector along the y-axis).

For example, a vector **r = [3, 2]** can be written as a combination of these basis vectors:
- **r = 3i + 2j**

Let's visualize this.
"""))

# Plot the basis vectors and r = 3i + 2j
i = np.array([1, 0])  # Unit vector along the x-axis
j = np.array([0, 1])  # Unit vector along the y-axis
r = 3 * i + 2 * j  # 3i + 2j

plt.figure(figsize=(6, 6))
plt.quiver(0, 0, i[0], i[1], angles='xy', scale_units='xy', scale=1, color="red", label="Vector i")
plt.quiver(0, 0, j[0], j[1], angles='xy', scale_units='xy', scale=1, color="green", label="Vector j")
plt.quiver(0, 0, r[0], r[1], angles='xy', scale_units='xy', scale=1, color="blue", label="Vector r = 3i + 2j")

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title("Coordinate System and Vector r = 3i + 2j")

plt.show()
```

```python
# In[7]: Section 6 - Vector Subtraction
# Visualizing vector subtraction.

s = np.array([-1, 2])  # Vector s

r_minus_s = r - s  # Vector subtraction

display(Markdown("""
### Vector Subtraction

To subtract vectors, we add the negative of a vector. For example:
- **r - s = r + (-s)**

This can be done component-wise. Let's visualize this operation.
"""))

# Plot the vectors
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, 3, 2, angles='xy', scale_units='xy', scale=1, color="blue", label="Vector r")
plt.quiver(3, 2, -1, 2, angles='xy', scale_units='xy', scale=1, color="red", label="Vector s")
plt.quiver(0, 0, r_minus_s[0], r_minus_s[1], angles='xy', scale_units='xy', scale=1, color="purple", label="r - s")

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title("Vector Subtraction (r - s)")

plt.show()
```
