```python
# 1. Price Discovery - Solving Simultaneous Equations

# Coefficients of the equations
A = np.array([[2, 3], [10, 1]])  # Coefficients matrix
B = np.array([8, 13])  # Result vector (costs)

# Solving the system of equations: Ax = B
prices = np.linalg.solve(A, B)

# Print the result: prices of apples and bananas
apple_price, banana_price = prices
print(f"Price of one apple: {apple_price:.2f} Euros")
print(f"Price of one banana: {banana_price:.2f} Euros")
```
