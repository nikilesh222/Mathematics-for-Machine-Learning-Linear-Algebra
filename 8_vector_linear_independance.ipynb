```python
import numpy as np

def check_linear_independence(vectors):
    # Convert the list of vectors to a numpy array (matrix form)
    matrix = np.column_stack(vectors)
    
    # Calculate the determinant of the matrix
    determinant = np.linalg.det(matrix)
    
    # If the determinant is zero, the vectors are linearly dependent
    if determinant == 0:
        return "The vectors are linearly dependent."
    else:
        return "The vectors are linearly independent."

# Example Vectors in 3D
vector_1 = np.array([1, 2, 3])
vector_2 = np.array([2, 4, 6])
vector_3 = np.array([3, 6, 9])

# Checking if these vectors are linearly independent or dependent
result = check_linear_independence([vector_1, vector_2, vector_3])
print(result)
```

```python
# Vectors representing edges of the pyramid
vector_1 = np.array([1, 0, 0])  # x-axis edge
vector_2 = np.array([0, 1, 0])  # y-axis edge
vector_3 = np.array([0, 0, 1])  # z-axis edge

# Checking linear independence to determine if the pyramid can exist in 3D space
result = check_linear_independence([vector_1, vector_2, vector_3])
print(result)
```

```python
# Forces acting on the object
force_1 = np.array([2, 1, 0])
force_2 = np.array([1, 3, 0])
force_3 = np.array([-1, -4, 0])

# Checking if the forces are linearly independent
result = check_linear_independence([force_1, force_2, force_3])
print(result)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to plot 3D vectors
def plot_vectors(vectors, labels, title):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    # Plot each vector
    for i, vec in enumerate(vectors):
        ax.quiver(0, 0, 0, vec[0], vec[1], vec[2], label=labels[i], arrow_length_ratio=0.1)
    
    # Set plot limits
    ax.set_xlim([0, 4])
    ax.set_ylim([0, 4])
    ax.set_zlim([0, 4])
    
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    
    # Title and legend
    ax.set_title(title)
    ax.legend()

    plt.show()

# Vectors for linearly independent example
v1 = np.array([1, 0, 0])
v2 = np.array([0, 1, 0])
v3 = np.array([0, 0, 1])

# Vectors for linearly dependent example
v4 = np.array([1, 0, 0])
v5 = np.array([2, 0, 0])
v6 = np.array([3, 0, 0])

# Plotting linearly independent vectors
plot_vectors([v1, v2, v3], ['v1', 'v2', 'v3'], 'Linearly Independent Vectors')

# Plotting linearly dependent vectors
plot_vectors([v4, v5, v6], ['v4', 'v5', 'v6'], 'Linearly Dependent Vectors')
```
