```python
import numpy as np
import matplotlib.pyplot as plt

# Define a matrix that scales by factors a and d
A = np.array([[2, 0],
              [0, 3]])

# Define the unit square
unit_square = np.array([[0, 0], 
                        [1, 0], 
                        [1, 1], 
                        [0, 1], 
                        [0, 0]])

# Apply the transformation (matrix multiplication)
transformed_square = unit_square.dot(A.T)

# Plotting the original and transformed square
plt.figure(figsize=(6,6))
plt.plot(unit_square[:,0], unit_square[:,1], label="Original Unit Square", color="blue")
plt.plot(transformed_square[:,0], transformed_square[:,1], label="Transformed Square", color="red")

# Adding details
plt.xlim(-1, 3)
plt.ylim(-1, 4)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title("Scaling Transformation of Unit Square")
plt.grid(True)
plt.show()

img_path = "./assets/images/16_7.png"
show_image(img_path, 200)
```

```python
img_path = "./assets/images/16_8.png"
show_image(img_path, 200)

# Define a more general matrix with shear and rotation
A_general = np.array([[2, 1],
                     [1, 2]])

# Apply the transformation to the unit square
transformed_square_general = unit_square.dot(A_general.T)

# Plotting the original and transformed shape
plt.figure(figsize=(6,6))
plt.plot(unit_square[:,0], unit_square[:,1], label="Original Unit Square", color="blue")
plt.plot(transformed_square_general[:,0], transformed_square_general[:,1], label="Transformed Shape", color="green")

# Adding details
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title("Transformation of Unit Square (General 2x2 Matrix)")
plt.grid(True)
plt.show()

# Calculate the determinant
det = np.linalg.det(A_general)
print(f"Determinant of A: {det}")

img_path = "./assets/images/16_9.png"
show_image(img_path, 200)
```

```python
img_path = "./assets/images/16_10.png"
show_image(img_path, 200)

# Check if the matrix is invertible
if det != 0:
    A_inv = np.linalg.inv(A_general)
    print(f"Inverse of A:\n{A_inv}")
else:
    print("Matrix is singular and does not have an inverse.")
```

```python
img_path = "./assets/images/16_11.png"
show_image(img_path, 200)

# Example of linearly independent vectors
A_independent = np.array([[2, 1],
                          [1, 3]])

# Example of linearly dependent vectors
A_dependent = np.array([[2, 4],
                        [1, 2]])

# Compute and print the determinants
det_independent = np.linalg.det(A_independent)
det_dependent = np.linalg.det(A_dependent)

print(f"Determinant of linearly independent matrix: {det_independent}")
print(f"Determinant of linearly dependent matrix: {det_dependent}")
```

```python
img_path = "./assets/images/16_12.png"
show_image(img_path, 200)

from mpl_toolkits.mplot3d import Axes3D

# Define a 3x3 matrix that maps 3D space to a plane
A_3d = np.array([[1, 1, 0],
                 [0, 1, 1],
                 [1, 0, 1]])

# Define a set of 3D points (unit cube)
points = np.array([[0, 0, 0], [1, 0, 0], [0, 1, 0], [0, 0, 1]])

# Apply the transformation
transformed_points = points.dot(A_3d.T)

# Plotting the original and transformed points in 3D space
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points[:,0], points[:,1], points[:,2], color='blue', label="Original Points")
ax.scatter(transformed_points[:,0], transformed_points[:,1], transformed_points[:,2], color='green', label="Transformed Points")

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.title("Transformation of 3D Unit Cube to a Plane")
plt.show()

# Calculate the determinant of the 3x3 matrix
det_3d = np.linalg.det(A_3d)
print(f"Determinant of 3x3 matrix: {det_3d}")
```
